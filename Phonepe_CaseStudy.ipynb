Task 3: Data Quality Checks


3.1: Ensure data consistency across state and district levels


```python
# From "District_Txn_and_Users" dataset: For each state, calculate the total number of transactions, total transaction amount, and total registered users.
From_District_Data = District_Txn_and_Users.groupby("State")[["Transactions", "Amount (INR)", "Registered Users"]].sum().reset_index()

# From "State_Txn_and_users" For each state, calculate the total number of transactions, total transaction amount, and total registered users.
From_state_data = State_Txn_and_users.groupby("State")[["Transactions", "Amount (INR)", "Registered Users"]].sum().reset_index()

# From the above values,Comparing the results with the corresponding values at the state level and district level to ensure they match.
comparison = pd.merge(From_District_Data,From_state_data, on="State", suffixes = ("_District","_State"))

# Rearraging the columns for better comparison
comparison = comparison[
    [
        "State",
        "Transactions_District", "Transactions_State",
        "Amount (INR)_District", "Amount (INR)_State",
        "Registered Users_District", "Registered Users_State"
    ]
]

# Display any discrepancies found between the district-level and state-level data.
discripancies = comparison[(comparison["Transactions_District"] != comparison["Transactions_State"]) |
(comparison["Amount (INR)_District"] != comparison["Amount (INR)_State"]) |
((comparison["Registered Users_District"] != comparison["Registered Users_State"]))
]

if discripancies.empty:
    print("No discrepancies found — state and district data match perfectly!")
else:
    print("Discrepancies found between state-level and district-level data:\n")
    print(discripancies)
```

4.1: Ratio of users to population by state


```python
# Merge the State_Txn and Users dataset with the District Demographics dataset to calculate the ratio of registered users to the population for each state. Display the results in a tabular format.
df = pd.merge(State_Txn_and_users, District_Demographics, on="State")
df["Ratio"] = df["Registered Users"] / df["Population"]

filter = df[["State", "Ratio"]]
ans = filter.groupby("State")["Ratio"].mean().reset_index()

# Create a column chart depicting the ratio of users to population by state.
plt.figure(figsize=(16,7))
plt.bar(ans["State"], ans["Ratio"], color = "Red")
plt.xlabel("State")
plt.ylabel("Ratio")
plt.title("Ratio of users to population by state.")
plt.xticks(rotation=90)
plt.show()

print("\n\nInsights:-\n-Telangana, Maharashtra, and Karnataka have the highest ratios of users to population, showing strong digital penetration in these states.\n-Regions with high ratios are best for launching new digital services, while those with low ratios have potential for further growth.\n-States like Delhi, Goa, and Ladakh have much lower ratios, meaning fewer users compared to their population.")
```

4.2: Correlate population density with transaction volume


```python
# Merge the District_Txn and Users dataset with the District Demographics dataset.
# Calculate the correlation between population density and transaction volume.
# Create a scatter plot to visualize the correlation between population density and transaction volume.

df = pd.merge(District_Txn_and_Users, District_Demographics, on="State")
proper = df.groupby("State")[["Density", "Transactions"]].agg({"Density": "mean", "Transactions": "sum"}).reset_index()
correl = proper["Density"].corr(proper["Transactions"])

print(f'Correlation between population density and transaction volume is {correl}.\n\n')

plt.figure(figsize=(16,6))
plt.scatter(proper["Density"], proper["Transactions"], color= "Purple")
plt.xlabel("Density")
plt.ylabel("Transactions")
plt.title("correlation between population density and transaction volume")
plt.show()

print("\n\nThe graph shows there is almost no link between population density and transaction volume; dense states do not always have more transactions.\n\n")
```
